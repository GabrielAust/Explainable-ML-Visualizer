# K-Means Demo

Run k-means clustering with console visualization export enabled.

```python
from data.dataset_generators import generate_blob_data
from models.kmeans import KMeans
from visualizations.common import VisualizationConfig
from visualizations.kmeans_viz import KMeansConsoleVisualizer

points, _ = generate_blob_data(samples=90, classes=3, class_overlap=0.2, noise=0.05)
config = VisualizationConfig(export_dir="exports/kmeans", export_png=True, export_gif=True)
visualizer = KMeansConsoleVisualizer(config)
model = KMeans(clusters=3, iterations=8)
visualizer.announce(clusters=3, iterations=8, dataset="Blob dataset")
state = model.fit(points, on_step=visualizer.update)
visualizer.summarize(state)
```

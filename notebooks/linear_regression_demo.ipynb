# Linear Regression Demo

Train a simple linear regression model and optionally export console frames as PNGs/GIFs.

```python
from data.dataset_generators import generate_linear_data
from models.linear_regression import LinearRegressionGD
from visualizations.common import VisualizationConfig
from visualizations.linear_regression_viz import LinearRegressionConsoleVisualizer

features, targets = generate_linear_data(slope=3.5, intercept=1.2, samples=40, noise=0.05)
config = VisualizationConfig(export_dir="exports/linear_regression", export_png=True, export_gif=True)
visualizer = LinearRegressionConsoleVisualizer(config)
model = LinearRegressionGD(learning_rate=0.1, iterations=30)
visualizer.announce(learning_rate=0.1, iterations=30)
state = model.fit(features, targets, on_step=visualizer.update)
visualizer.summarize(state)
```
